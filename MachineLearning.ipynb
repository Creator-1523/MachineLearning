Linear Regression


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/py/ML/1_linear_reg/homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel("area")
plt.ylabel("Price(US)")
plt.scatter(df.area,df.price,color='blue',marker='+')
```

```python
new_df=df.drop('price',axis='columns')
new_df
```

```python
price=df.price
price
```

Create Linear Regression Object

```python
reg=linear_model.LinearRegression()
reg.fit(new_df,df.price)   #reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#y=m*x+b
135.78767123*3300+180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('area',fontsize=20)
plt.ylabel('price',fontsize=20)
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```

```python
reg.predict([[5000]])
```

```python
area_df = pd.read_csv("/content/py/ML/1_linear_reg/areas.csv")
area_df.head(3)
```

Predict canada's per capita income in year 2020. 

```python
cdf=pd.read_csv('/content/canada_per_capita_income.csv')
cdf
```

```python
new_cdf=pd.DataFrame(cdf)
new_cdf.rename(columns={'per capita income (US$)':'capita'},inplace=True)
new_cdf
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(cdf.year,cdf.capita,color='red',marker='+')
```

```python
new_cdf=cdf.drop('capita',axis='columns')
new_cdf
```

```python
capita=cdf.capita
capita
```

```python
creg=linear_model.LinearRegression()
creg.fit(new_cdf,capita)
```

```python
creg.predict([[2020]])
```

```python
%matplotlib inline
plt.xlabel('year',fontsize=20)
plt.ylabel('capita',fontsize=20)
plt.scatter(cdf.year,cdf.capita,color='red',marker='+')
plt.plot(cdf.year,creg.predict(cdf[['year']]),color='blue')
```

LinearRegression Using Two Variables

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
tdf=pd.read_csv('/content/py/ML/2_linear_reg_multivariate/homeprices.csv')
tdf
```

```python
import math
median_bedrooms=math.floor(tdf.bedrooms.median())

median_bedrooms
```

```python
tdf.bedrooms=tdf.bedrooms.fillna(median_bedrooms)
tdf
```

```python
treg=linear_model.LinearRegression()
treg.fit(tdf[['area','bedrooms','age']],tdf.price)
```

```python
treg.coef_
```

```python
treg.intercept_
```

```python
treg.predict([[3000,4,15]])
```

```python
 112.06244194*3000+23388.88007794*4-3231.71790863*15+221323.00186540396
```

```python
treg.predict([[2500,4,5]])
```
